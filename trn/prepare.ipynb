```python
import pandas as pd
import numpy as np
import os
import math
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle
```

```python
data = pd.read_csv("./TRNcod.csv").set_index("INDEX")

data.head()
```

```python
data.shape
```

```python
data1 = data[ data["IND_BOM_1_1"] == 1 ]

data1.head()
```

```python
data1.shape
```

```python
data2 = data[ data["IND_BOM_1_2"] == 1]

data2.head()
```

```python
data2.shape
```

```python
def create_x_y(df):
    x = df.iloc[:, :243].to_numpy()
    y = df.iloc[:, 243:].to_numpy()

    assert x.shape[0] == y.shape[0]
    assert x.shape[0] == df.shape[0]

    return x, y

x1, y1 = create_x_y(data1)
x2, y2 = create_x_y(data2)
```

```python
def train_test_validation_split(x, y):
    x_train, x_temp, y_train, y_temp = train_test_split(x, y, train_size=.5)

    x_test, x_val, y_test, y_val = train_test_split(x_temp, y_temp, train_size=.5)

    assert x_train.shape[0] == y_train.shape[0]
    assert x_val.shape[0] == y_val.shape[0]    
    assert x_test.shape[0] == y_test.shape[0]

    assert x_train.shape[0] == math.floor(x.shape[0] * .5)
    assert x_val.shape[0] == math.ceil(x.shape[0] * .25)
    assert x_test.shape[0] == math.floor(x.shape[0] * .25)

    return x_train, x_val, x_test, y_train, y_val, y_test

x1_train, x1_val, x1_test, y1_train, y1_val, y1_test = train_test_validation_split(x1, y1)
x2_train, x2_val, x2_test, y2_train, y2_val, y2_test = train_test_validation_split(x2, y2)
```

```python
# upsample train and validation data

difference_train = x1_train.shape[0] - x2_train.shape[0]

x2_train_upsampled = np.concatenate((x2_train, x2_train[:difference_train]))
y2_train_upsampled = np.concatenate((y2_train, y2_train[:difference_train]))

difference_val = x1_val.shape[0] - x2_val.shape[0]

x2_val_upsampled = np.concatenate((x2_val, x2_val[:difference_val]))
y2_val_upsampled = np.concatenate((y2_val, y2_val[:difference_val]))

assert x2_train_upsampled.shape[0] == y2_train_upsampled.shape[0]
assert x2_train_upsampled.shape == x1_train.shape
assert y2_train_upsampled.shape == y1_train.shape

assert x2_val_upsampled.shape[0] == y2_val_upsampled.shape[0]
assert x2_val_upsampled.shape == x1_val.shape
assert y2_val_upsampled.shape == y1_val.shape
```

```python
# join datasets

x_train = np.concatenate((x1_train, x2_train_upsampled))
y_train = np.concatenate((y1_train, y2_train_upsampled))

x_val = np.concatenate((x1_val, x2_val_upsampled))
y_val = np.concatenate((y1_val, y2_val_upsampled))

x_test = np.concatenate((x1_test, x2_test))
y_test = np.concatenate((y1_test, y2_test))

print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
print(x_test.shape,y_test.shape)
```

```python
x_train, y_train = shuffle(x_train, y_train)
x_val, y_val = shuffle(x_val, y_val)

print(x_train.shape, y_train.shape)
print(x_val.shape, y_val.shape)
```

```python
## save data
os.mkdir("data")

np.savez("data/train.npz", x=x_train, y=y_train)
np.savez("data/val.npz", x=x_val, y=y_val)
np.savez("data/test.npz",x=x_test,y=y_test)
```
